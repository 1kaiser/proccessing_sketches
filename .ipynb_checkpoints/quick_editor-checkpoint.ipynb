```python
import re
import os
import shutil
from pathlib import Path

def process_pde_files(pde_file, output_dir):
    """
    Processes a list of PDE files, making the specified size and font replacements,
    and saving the updated files in individual folders within a specified directory.

    Args:
        pde_files: A list of paths to PDE files (strings or Path objects).
        output_dir: The directory to create subfolders and save the updated files.
    """

    with open(pde_file, 'r') as f:
        pde_content = f.read()

    # Replace size() using regex with capture groups
    pde_content = re.sub(r"PFont", r"//PFont", pde_content)
    pde_content = re.sub(r"PImage", r"//PImage", pde_content)
    pde_content = re.sub(r"img\.width", r"7860", pde_content)
    pde_content = re.sub(r"img\.height", r"4320", pde_content)
    pde_content = re.sub(r"img", r"//img", pde_content)

    pde_content = re.sub(r"image", r"//image", pde_content)

    pde_content = re.sub(r"background", r"//background", pde_content)
    pde_content = re.sub(r"size", r"size( 7860, 4320);//size", pde_content)

    pde_content = re.sub(r"size\( (\d+), (\d+)\);", r"size( \1/3, \2/3);", pde_content)

    # Comment out font lines using regex
    pde_content = re.sub(r"font", r"//font", pde_content)
    pde_content = re.sub(r"textFont", r"//textFont", pde_content)


    file_name = Path(pde_file).stem
    output_folder = Path(output_dir) / file_name
    output_folder.mkdir(parents=True, exist_ok=True)
    output_file_path = output_folder / Path(pde_file).name
    with open(output_file_path, 'w') as outfile:
        outfile.write(pde_content)

    print(f"Processed and saved: {output_file_path}")

output_dir = "/home/kaiser/sketchbook/all/proccessing_sketches/test" 
process_pde_files(pde_files[208], output_dir)  # Process all files found
```

```python
import os
import fnmatch
from pathlib import Path

def find_and_sort_pngs(directory):
    """
    Finds all PNG files within a directory and its subdirectories, 
    filters out folders without PNG/TIFF files, and sorts the filenames.

    Args:
        directory: The directory to search.

    Returns:
        A list of paths to PNG files, sorted alphabetically by folder name then filename.
        Returns an empty list if no PNG files are found or if there is an error.
    """
    
    image_list = []
    try:
        for root, _, files in os.walk(directory):
            png_files = []
            for filename in files:
                if fnmatch.fnmatch(filename, '*.png') or fnmatch.fnmatch(filename, '*.tiff'):
                    png_files.append(os.path.join(root, filename))
            
            if png_files:  # Only proceed if the folder contains at least one PNG
                sorted_pngs = sorted(png_files)  # Sorts by filename
                image_list.extend(sorted_pngs)
    except Exception as e:
        print(f"An error occurred: {e}")
        return []
    
    return image_list
# Example Usage
top_level_directory = "/home/kaiser/sketchbook/all/proccessing_sketches/test" #Replace with your directory

image_paths = find_and_sort_pngs(top_level_directory)

if image_paths:
    for image_path in image_paths:
        print(image_path)
else:
    print("No PNG or TIFF files found in the specified directory or subdirectories.")
```

```python
import os
import fnmatch
from pathlib import Path
import re

def find_and_sort_pngs(directory):
    """Finds PNG files, filters folders, and sorts by folder name."""
    image_data = []
    for root, _, files in os.walk(directory):
        png_files = [f for f in files if f.endswith('.png')]  # Simplified PNG check
        if png_files:
            folder_name = os.path.basename(root)
            for png_file in png_files:
                file_path = os.path.join(root, png_file)
                image_data.append((folder_name, file_path))

    image_data.sort(key=lambda x: x[0])  # Sort by folder name
    sorted_image_paths = [x[1] for x in image_data]
    return sorted_image_paths

def replace_strings_in_list(input_list, old_string, new_string):
    """Replaces strings and spaces."""
    return [s.replace(old_string, new_string).replace(" ", "%20") for s in input_list]

def replace_table_in_readme(readme_path, new_list):
    """Replaces the table in readme.md."""
    try:
        with open(readme_path, "r") as f:
            readme_content = f.read()

        table_start = r"test results\n\| outputs \| outputs \| outputs \|\n\|---\|---\|---\|\n"
        match = re.search(table_start, readme_content)

        if not match:
            print("Table not found in readme.md.  Creating a new table.") # Improved message
            table_end_index = len(readme_content)  # Insert at end if not found
            new_table_string = f"{table_start}"  # Add the table header if creating new

        else:
            table_end_index = match.end()
            new_table_string = ""

        for i in range(0, len(new_list), 3):
            row_parts = [f"![]({url})" for url in new_list[i:i + 3]]
            row = "|".join(row_parts)
            new_table_string += f"|{row}|\n"

        new_readme_content = readme_content[:table_end_index] + new_table_string

        with open(readme_path, "w") as f:
            f.write(new_readme_content)

        print("Table in README.md updated successfully.")

    except FileNotFoundError:
        print(f"Error: readme.md file not found at {readme_path}")
    except Exception as e:
        print(f"An error occurred: {e}")



# Example Usage (replace with your actual values)
top_level_directory = "/home/kaiser/sketchbook/all/proccessing_sketches/test"
readme_path = "README.md"  
old_string = "/home/kaiser/sketchbook/all/proccessing_sketches"
new_string = "https://github.com/1kaiser/proccessing_sketches/blob/main"

image_paths = find_and_sort_pngs(top_level_directory)

if image_paths:
    new_list = replace_strings_in_list(image_paths, old_string, new_string)
    replace_table_in_readme(readme_path, new_list)
else:
    print("No PNG files found, skipping table update.")

image_paths
```
